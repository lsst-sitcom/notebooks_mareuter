```python
from astropy.io import fits
from astropy.utils.data import download_file
from astropy.visualization import astropy_mpl_style
import matplotlib.pyplot as plt
import numpy as np

from lsst.sitcom.mareuter.site_lfa import get_lfa

plt.style.use(astropy_mpl_style)
```

```python
lfa = get_lfa()
url = f"{lfa.endpoint_url}/{lfa.bucket}/Electrometer:103/fits/2025/09/10/EM103_O_20250910_000011.fits"
```

```python
image_file = download_file(url)
image_data = fits.getdata(image_file, ext=1)
print(image_data.shape[0])
```

```python
x = np.array([i[1] for i in image_data])
y = np.array([i[0] for i in image_data])
```

```python
plt.plot(x, y)
```
